```python
import networkx as nx
import csv
from networkx.algorithms import bipartite
import random
```

```python
# extract user and artist nodes from csv File

with open('Node_Labels.csv','r') as node_csv:

  node_reader = csv.reader(node_csv)
  nodes = [n for n in node_reader][1:]


users = [n[1] for n in nodes]
print(users)



with open('art.csv','r') as node_csv:

  node_reader = csv.reader(node_csv)
  nodes = [n for n in node_reader][1:]


artists = [n[1] for n in nodes]
print(artists)

```

```python
# Make Graph object and then add two different types of nodes to it (User and Artist)
G = nx.DiGraph()
G.add_nodes_from(users, bipartite=0)
G.add_nodes_from(artists,bipartite=1)

users = []
artists = []
count = 0

for node in G.nodes():

  count = count + 1

  if count > 1134 :

    artists.append(node)
  else:
    users.append(node)


print(users)
print(artists)
```

```python
# Add Edges to the Graph
with open('user-artist.csv','r') as edge_csv:

  edge_reader = csv.reader(edge_csv)
  edges = [tuple(e) for e in edge_reader][1:]

print(edges)
new_edges = []
edges_new = []
for edge in edges:
  new_edges += [tuple([edge[0],edge[1],edge[2]])]           

for i in range(len(new_edges)):
  edge1=int(new_edges[i][0])
  edge2 =int(new_edges[i][1])
  edge1 = users[edge1-1]
  edge2 = artists[edge2]
  weight = int(new_edges[i][2])
  edges_new.append(tuple([edge1,edge2,{'weight':weight}]))


G.add_edges_from(edges_new)
  
```

```python
# Draw Bipartite Layout
pos = nx.bipartite_layout(G,users)
nx.draw(G,pos, with_labels=True)
```

```python
# Algoritm for finding common artist

similarity_index = []
final_list = []

for user in users:

  user_adjacency_list = []
  user_dictionary = {}

  for i in range(len(artists)):
    artist = artists[i]

    if artist not in user_adjacency_list:

      if G.has_edge(user,artist):
        weight = edges_new[i][2]['weight']
        user_adjacency_list.append(tuple((artist,weight)))
    
    user_dictionary['user'] = user
    user_adjacency_list = sorted(user_adjacency_list, key=lambda x: x[1],reverse=True)
    user_dictionary['adjacency_list'] = user_adjacency_list
    
  final_list.append(user_dictionary)

print(final_list)
```
